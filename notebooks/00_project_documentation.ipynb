<div style="background:#f8fafc;border:1px solid #e2e8f0;border-radius:0 0 10px 10px;padding:24px;">

  <h3 style="color:#0f172a;margin:20px 0 12px;font-size:1em;">🎯 Business Objectives</h3>
  <div style="display:grid;grid-template-columns:1fr 1fr;gap:12px;margin-bottom:20px;">

      <strong style="color:#1e40af;font-size:.88em;">Demand Forecasting</strong>
      <p style="color:#475569;font-size:.85em;margin:4px 0 0 0;">Predict weekly demand for 10 industrial SKUs one week ahead, enabling proactive production planning.</p>
    </div>
    <div style="background:white;border-left:3px solid #10b981;padding:12px 16px;border-radius:0 8px 8px 0;">
      <strong style="color:#065f46;font-size:.88em;">Supply Risk Detection</strong>
      <p style="color:#475569;font-size:.85em;margin:4px 0 0 0;">Automatically detect demand spikes, drops, and supplier reliability issues before they become problems.</p>
    </div>
    <div style="background:white;border-left:3px solid #f59e0b;padding:12px 16px;border-radius:0 8px 8px 0;">
      <strong style="color:#92400e;font-size:.88em;">Explainability</strong>
      <p style="color:#475569;font-size:.85em;margin:4px 0 0 0;">Provide SHAP-based explanations so supply chain managers understand why the model made each prediction.</p>

  <p style="color:#475569;line-height:1.8;font-size:.88em;">
    The simulation covers a 2-year period (104 weeks) with 10 industrial products across mechanical and electronic categories,
    5 industrial clients (automotive, construction, energy, medical, aerospace), and 3 raw material suppliers from Germany, Taiwan, and Spain.

  <h2 style="color:white;margin:0;font-size:1.1em;font-weight:700;">6. ML Model (Notebook 03)</h2>
</div>

<div style="background:#f8fafc;border:1px solid #e2e8f0;border-radius:0 0 10px 10px;padding:24px;">

  <h3 style="color:#0f172a;font-size:.95em;margin:0 0 12px 0;">6.1 Algorithm: XGBoost</h3>
  <p style="color:#475569;line-height:1.8;font-size:.88em;margin-bottom:16px;">
    XGBoost (Extreme Gradient Boosting) builds an ensemble of decision trees sequentially, where each tree corrects the errors of the previous ones. It was chosen for this project because it handles mixed variable types naturally, is robust to outliers (plant shutdowns, holiday spikes), learns feature interactions automatically, and is fully compatible with SHAP for explainability.
  </p>

  <h3 style="color:#0f172a;font-size:.95em;margin:0 0 12px 0;">6.2 Walk-Forward Validation</h3>
  <div style="background:#fff7ed;border-left:4px solid #f59e0b;padding:14px 18px;border-radius:0 8px 8px 0;margin-bottom:16px;">
    <strong style="color:#92400e;font-size:.88em;">Why not a random train/test split?</strong>
    <p style="color:#78350f;font-size:.83em;margin:6px 0 0 0;">
      In time series, a random split would allow the model to train on week 80 and predict week 30 — impossible in production. Walk-forward validation simulates real deployment: train on weeks 0-39, predict week 40; train on weeks 0-40, predict week 41; and so on. The metrics reflect actual production performance.
    </p>
  </div>
  <div style="background:#1e1e1e;border-radius:8px;padding:16px;font-family:monospace;font-size:.8em;color:#a8ff78;line-height:1.9;margin-bottom:16px;">
    Fold 1:  Train [weeks 0-39]  → Predict week 40<br>
    Fold 2:  Train [weeks 0-40]  → Predict week 41<br>
    Fold 3:  Train [weeks 0-41]  → Predict week 42<br>
    ...<br>
    Fold N:  Train [weeks 0-T-1] → Predict week T<br>
    Final metrics = average across all folds
  </div>

  <h3 style="color:#0f172a;font-size:.95em;margin:0 0 12px 0;">6.3 Hyperparameters</h3>

  <h3 style="color:#0f172a;font-size:.95em;margin:20px 0 12px 0;">6.5 SHAP Explainability</h3>
  <p style="color:#475569;line-height:1.8;font-size:.88em;">
    SHAP (SHapley Additive exPlanations) assigns each feature a contribution value for every individual prediction. Unlike global feature importance, SHAP explains why the model made a specific prediction for a specific product in a specific week — making the model fully interpretable to supply chain managers who need to trust and act on its output.
  </p>
</div>

  <h2 style="color:white;margin:0;font-size:1.1em;font-weight:700;">9. Key Design Decisions</h2>
</div>

<div style="background:#f8fafc;border:1px solid #e2e8f0;border-radius:0 0 10px 10px;padding:24px;">
  <div style="display:flex;flex-direction:column;gap:14px;">

    <div style="background:white;border:1px solid #e2e8f0;border-radius:10px;padding:16px 20px;">
      <div style="display:flex;gap:12px;align-items:flex-start;">
        <span style="font-size:1.4em;">🌳</span>
        <div>
          <strong style="color:#0f172a;">XGBoost over Prophet or LSTM</strong>
          <p style="color:#475569;font-size:.85em;margin:6px 0 0 0;">Prophet handles seasonality well but is limited to univariate data. LSTM requires much more data and is harder to explain. XGBoost handles 70+ mixed variables naturally, trains in seconds, and produces SHAP values — the best tradeoff for this use case.</p>
        </div>
      </div>
    </div>

    <div style="background:white;border:1px solid #e2e8f0;border-radius:10px;padding:16px 20px;">
      <div style="display:flex;gap:12px;align-items:flex-start;">
        <span style="font-size:1.4em;">⏱️</span>
        <div>
          <strong style="color:#0f172a;">Walk-forward over random split</strong>
          <p style="color:#475569;font-size:.85em;margin:6px 0 0 0;">A random split in time series data produces optimistic metrics that don't reflect production reality. Walk-forward validation simulates every Monday morning deployment — the only way to get honest performance estimates.</p>
        </div>
      </div>
    </div>

    <div style="background:white;border:1px solid #e2e8f0;border-radius:10px;padding:16px 20px;">
      <div style="display:flex;gap:12px;align-items:flex-start;">
        <span style="font-size:1.4em;">🔢</span>
        <div>
          <strong style="color:#0f172a;">Multiplicative demand model</strong>
          <p style="color:#475569;font-size:.85em;margin:6px 0 0 0;">Demand factors are multiplied together rather than added. This is more realistic: a holiday during a peak season is worse than either alone, and a promotion during an economic downturn has a dampened effect. Multiplicative models capture these compound effects naturally.</p>
        </div>
      </div>
    </div>

    <div style="background:white;border:1px solid #e2e8f0;border-radius:10px;padding:16px 20px;">
      <div style="display:flex;gap:12px;align-items:flex-start;">
        <span style="font-size:1.4em;">🔍</span>
        <div>
          <strong style="color:#0f172a;">SHAP for explainability</strong>
          <p style="color:#475569;font-size:.85em;margin:6px 0 0 0;">In supply chain, managers need to trust and understand model predictions before acting on them. SHAP transforms XGBoost from a black box into a transparent system — every prediction comes with a ranked list of which variables drove it up or down.</p>
        </div>
      </div>
    </div>

    <div style="background:white;border:1px solid #e2e8f0;border-radius:10px;padding:16px 20px;">
      <div style="display:flex;gap:12px;align-items:flex-start;">
        <span style="font-size:1.4em;">⚡</span>
        <div>
          <strong style="color:#0f172a;">Singleton model loading</strong>
          <p style="color:#475569;font-size:.85em;margin:6px 0 0 0;">The XGBoost model is loaded once at API startup via FastAPI's lifespan context manager and shared across all requests. This reduces response time from ~1s (loading from disk each time) to ~1ms (inference only).</p>
        </div>
      </div>
    </div>

  </div>
